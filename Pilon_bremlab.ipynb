```python
import os
import pandas as pd
```

```python
pilon_files = "pilon_changes_renamed/"
# original_pilon_files = os.listdir(pilon_files)
```

```python
from glob import glob
    
pima_files = glob(pilon_files + "PIMA*")
```

```python
x = pima_files[0]
```

```python
x
```

```python
x.split(".")[0].split("/")[1]
```

```python
for i in pima_files:
    changes = pd.read_csv(i, header=None, sep=' ')
    changes.columns = ["orig_coord", "new_coord", "orig_seq", "new_seq"]
    length_orig_seq = changes.orig_seq.apply(lambda x: len(x))
    changes["length_orig_seq"] = length_orig_seq
    length_new_seq = changes.new_seq.apply(lambda x: len(x))
    changes["length_new_seq"]= length_new_seq
    N_count = changes.orig_seq.apply(lambda x: x.count('N'))
    changes = changes[(changes.length_orig_seq > 3000) & (changes.length_new_seq < 500) & (N_count < 10)]
    
    changes.to_csv(os.path.join("Filtered_pima/", i.split(".")[0].split("/")[1] + "_deletions.csv"), index=False)
```
